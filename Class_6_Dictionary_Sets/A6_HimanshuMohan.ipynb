```python
# how_many.py
#
#  function that takes a list with an integer and a noun, and 
#  returns a sentence describing the situation
# Usage:
#    % python how_many.py one input parameter of list with an integer and a noun
#
# Himanshu Mohan, Oct 14, 2019

from typing import List

def how_many(lst: List) -> str:
    "How many are there?"
    if lst==[]:
        print('empty list')

    
    if lst[0] >1 :
         return('There are '+str(lst[0])+' '+str(lst[1])+"s")
      
    else:
          return('There is '+str(lst[0])+' '+str(lst[1]))

how_many([1, "TF"])
```

```python
# string_to_int.py
#
#  a function that takes a string of comma-separated integers and returns 
#  a list of the integers
# Usage:
#    % python string_to_int.py one input parameter of string and return int list
#
# Himanshu Mohan, Oct 15, 2019

from typing import List

def strings_to_ints(text: str) -> List[int]:
    "Take a comma-seperated list of integers and return a list of integers"
    if text=='':
        return[]
    try:
        str_lst=text.split(",")
        int_lst=[]
        for item in str_lst:
            int_lst.append(int(item))  
            
        return (int_lst)
    except:
            print('Invalid Integer Conversion for string- ', item)
            return()
                  
print (strings_to_ints('1,2,3,4,5'))
```

```python
# oxford_comma.py
#
#  a function which takes a list of words, and returns a string 
# with a properly punctuated sequence
# Usage:
#    % python oxford_commat.py one input parameter of list and return string
#
# Himanshu Mohan, Oct 15, 2019



from typing import List

def oxford_comma(lst: List[str]) -> str:
    "Take a list and apply the rules for Oxford Comma"
    if lst==[]:
        return ('')
        
    length=len(lst)
    string=''
    if length==1:
        string=lst[0]
    if length==2:
        string=lst[0]+' '+'and' +' '+ lst[1]
    if length>2:
        for index in range(length):
            if index<length-1:
                string=string+lst[index]+','+' '
            else:
                 string=string+'and' +' '+lst[length-1]

    
    return (string)

print(oxford_comma(['Snap', 'Crackle', 'Pop' ]))
```

```python
# parenthesis.py
#
#  a function is_valid_parenthesis() that takes a string and decides 
#  if it consists of valid nested parenthesis
# Usage:
#    % python parenthesis.py one input parameter of string and return True or False
#
# Himanshu Mohan, Oct 15, 2019

from typing import List

def is_valid_parenthesis(expression: str) -> bool:
    "Is this string a valid set of parenthesis?"
    if expression=='':
        return True
    open_list = ["[","{","("] 
    close_list = ["]","}",")"] 
    stack = [] 
    for ch in expression: 
        if ch in open_list: 
            stack.append(ch) 
        elif ch in close_list: 
            pos = close_list.index(ch) 
            if ((len(stack) > 0) and (open_list[pos] == stack[len(stack)-1])): 
                stack.pop() 
            else: 
                return False
    if len(stack) == 0: 
        return True
    else:
        return False
    
print(is_valid_parenthesis('([({[])])'))
```

```python
# Count_bigrams.py
#
# Count the frequency of each pair of bases in a sequence found in a Fasta File

from typing import Dict

def readFastaFile(fileName: str) -> str:
    "Read in a fasta file, return contents"

    try:
         with open(fileName,"r") as myfile:
            next(myfile) 
            data=myfile.read()
        
    except IOError:
        return ''
    return (data)

def countDigrams(fileName: str) -> Dict[str, int]:
    "Count the pairs in the file"

    text = readFastaFile(fileName)

    pairsCount: Dict[str, int] = {}

    length=len(text)

    count_AA=0
    count_AG=0
    count_AC=0
    count_AT=0
    count_GA=0
    count_GG=0
    count_GC=0
    count_GT=0
    count_CA=0
    count_CG=0
    count_CC=0
    count_CT=0
    count_TA=0
    count_TG=0
    count_TC=0
    count_TT=0
    for index in range(length):
         if index< length-1:
            
            if text[index:index+2]=='AA':
                 count_AA=count_AA+1;
            if text[index:index+2]=='AG':
                count_AG=count_AG+1;
            if text[index:index+2]=='AC':
                count_AC=count_AC+1; 
            if text[index:index+2]=='AT':
                count_AT=count_AT+1; 
            if text[index:index+2]=='GA':
                 count_GA=count_GA+1;
            if text[index:index+2]=='GG':
                count_GG=count_GG+1;
            if text[index:index+2]=='GC':
                count_GC=count_GC+1; 
            if text[index:index+2]=='GT':
                count_GT=count_GT+1; 
            if text[index:index+2]=='CA':
                 count_CA=count_CA+1;
            if text[index:index+2]=='CG':
                count_CG=count_CG+1;
            if text[index:index+2]=='CC':
                count_CC=count_CC+1; 
            if text[index:index+2]=='CT':
                count_CT=count_CT+1;
            if text[index:index+2]=='TA':
                 count_TA=count_TA+1;
            if text[index:index+2]=='TG':
                count_TG=count_TG+1;
            if text[index:index+2]=='TC':
                count_TC=count_TC+1; 
            if text[index:index+2]=='TT':
                count_TT=count_TT+1; 
    pairsCount={'AA':count_AA,'AG':count_AG, 'AC':count_AC,'AT':count_AT, 'GA':count_GA,'GG':count_GG, 'GC':count_GC,'GT':count_GT,'CA':count_CA,'CG':count_CG, 'CC':count_CC,'CT':count_CT, 'TA':count_TA,'TG':count_TG, 'TC':count_TC,'TT':count_TT}
    return pairsCount


def printDigrams(pairsCount: Dict[str, int]):
    "Print the digrams"

    bases = ['A', 'G', 'C', 'T']

    # Print the column headings
    print(' ', end=' ')
    for ch in bases:
        print(ch.rjust(7), end=' ')
    print()

    # Print the body of the table
    for ch1 in bases:
        print(ch1, end=' ')

        # Print a row of the table
        for ch2 in bases:
            digram = ch1 + ch2
            if (digram in pairsCount):
                count = pairsCount[digram]
            else:
                count = 0

            # Print count, with formating
            print(repr(count).rjust(7), end=' ')
        print()

pairsCount = countDigrams('/course/data/ecoli.fasta')
printDigrams(pairsCount)
```
